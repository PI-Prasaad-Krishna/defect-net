```python
import cv2
import matplotlib.pyplot as plt
import os

# Define the path to your processed data
base_path = os.path.join('..', 'datasets', 'processed', 'track_classification', 'train')

# Define paths to the two classes
defective_path = os.path.join(base_path, 'defective')
non_defective_path = os.path.join(base_path, 'non_defective')

# Get the first image file from each directory
defective_image_file = os.path.join(defective_path, os.listdir(defective_path)[0])
non_defective_image_file = os.path.join(non_defective_path, os.listdir(non_defective_path)[0])

# Read the images
defective_img = cv2.imread(defective_image_file)
non_defective_img = cv2.imread(non_defective_image_file)

# OpenCV loads images in BGR format, but Matplotlib displays in RGB. We need to convert.
defective_img_rgb = cv2.cvtColor(defective_img, cv2.COLOR_BGR2RGB)
non_defective_img_rgb = cv2.cvtColor(non_defective_img, cv2.COLOR_BGR2RGB)

# Display the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.title('Defective')
plt.imshow(defective_img_rgb)
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Non-Defective')
plt.imshow(non_defective_img_rgb)
plt.axis('off')

plt.show()
```

```python
import tensorflow as tf

# Define constants
IMG_SIZE = (256, 256)
BATCH_SIZE = 32

# Define the paths to your data directories
train_dir = os.path.join('..', 'datasets', 'processed', 'track_classification', 'train')
validation_dir = os.path.join('..', 'datasets', 'processed', 'track_classification', 'validation')
test_dir = os.path.join('..', 'datasets', 'processed', 'track_classification', 'test')

# Load the datasets
train_dataset = tf.keras.utils.image_dataset_from_directory(
    train_dir,
    label_mode='binary',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

validation_dataset = tf.keras.utils.image_dataset_from_directory(
    validation_dir,
    label_mode='binary',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

test_dataset = tf.keras.utils.image_dataset_from_directory(
    test_dir,
    label_mode='binary',
    image_size=IMG_SIZE,
    batch_size=BATCH_SIZE
)

# Print the class names that Keras found
print("Class names found:", train_dataset.class_names)
```
